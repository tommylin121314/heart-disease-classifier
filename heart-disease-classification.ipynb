```python
# Split data into x and y 
x = hd_df.drop("target", axis=1)
y = hd_df['target']
```

```python
# Split data into train and test sets
np.random.seed(42)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
# There are three models we want to try:
# 1. Logistic Regression
# 2. KNeighbors Classifier
# 3. Random Forest Classifier

# Create a dictionary containing all models
models = {"Logistic Regression": LogisticRegression(),
          "KNN": KNeighborsClassifier(),
          "Random Forest": RandomForestClassifier()}

# Create a method to fit and score our models:
def fit_and_score(models, x_train, x_test, y_train, y_test):
    """
    models: dictionary containing sklearn models
    x_train: non-labeled training set
    x_test: non-labeled testing set
    y_train: labeled training set
    y_test: labeled testing set
    """
    
    # Set random seed for consistent results
    np.random.seed(42)
    
    # Create dictionary to hold model scores
    model_scores = {}
    
    # loop through models
    for name, model in models.items():
        # Fit model
        model.fit(x_train, y_train)
        # Evalutate model
        model_scores[name] = model.score(x_test, y_test)
        
    return model_scores
```

```python
# Fit and score all models
model_scores = fit_and_score(models, x_train, x_test, y_train, y_test)

# Visualize the results using a bar graph
model_scores_df = pd.DataFrame(model_scores, index=['accuracy'])
model_scores_df.T.plot(kind="bar")
plt.xlabel("Model")
plt.ylabel("Accuracy")
plt.xticks(rotation=0);
```

```python
# Hyperparameter tuning for the KNN

train_scores = []
test_scores = []

# Create list of different values for n neighbors (a parameter in KNN)
neighbors = range(1, 21)

# Setup KNN
knn = KNeighborsClassifier()

# Iterate over list of neighbor values and fit the data using KNN with that number of neighbors
for i in neighbors:
    
    # Fit KNN with neighbor
    knn.set_params(n_neighbors=i)
    knn.fit(x_train, y_train)
    
    # Append scores to score arrays
    train_scores.append(knn.score(x_train, y_train))
    test_scores.append(knn.score(x_test, y_test))
```

```python
# Plot scores data for KNN
plt.plot(neighbors, train_scores, label="Train Scores")
plt.plot(neighbors, test_scores, label="Test Scores")
plt.xlabel("Number of Neighbors")
plt.ylabel("Accuracy")
plt.xticks(neighbors)
plt.legend();
```

```python
# Hyperparameter tuning of with RandomSearchCV

# Create hyperparameter grid for LogisticRegression
log_reg_grid = {"C": np.logspace(-4, 4, 20), 
                "solver": ["liblinear"]}

# Create hyperparameter grid for RandomForestClassifier
rf_grid = {"n_estimators": np.arange(10, 1000, 50),
           "max_depth": [None, 3, 5, 10],
           "min_samples_split": np.arange(2, 20, 2),
           "min_samples_leaf": np.arange(1, 20, 2)}
```

```python
# Tune LogisticRegression

np.random.seed(42)

# Setup random hyperparameter search for LogisticRegression
rs_log_reg = RandomizedSearchCV(LogisticRegression(), 
                                param_distributions=log_reg_grid, 
                                cv=5, 
                                n_iter=20, 
                                verbose=True)

# Fit random hyperparameter search model for LogisticRegression
rs_log_reg.fit(x_train, y_train)
```

```python
rs_log_reg.best_params_
```

```python
rs_log_reg.score(x_test, y_test)
```

```python
# Tune RandomForestClassifier

np.random.seed(42)

# Setup random hyperparameter search for RandomForestRegression
rf_rs = RandomizedSearchCV(RandomForestClassifier(),
                               param_distributions=rf_grid,
                               cv=5,
                               n_iter=20,
                               verbose=True)

rf_rs.fit(x_train, y_train)
```

```python
rf_rs.best_params_
```

```python
# Hyperparameter tuning with GridSearchCV

# Create hyperparameter grid for LogisticRegression
log_reg_grid = {"C": np.logspace(-4, 4, 30), 
                "solver": ["liblinear"]}

gs_log_reg = GridSearchCV(LogisticRegression(),
                          param_grid=log_reg_grid,
                          cv=5,
                          verbose=True)

gs_log_reg.fit(x_train, y_train)
```
